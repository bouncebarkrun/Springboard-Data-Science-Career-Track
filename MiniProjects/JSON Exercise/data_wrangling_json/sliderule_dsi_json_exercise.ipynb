Using data in file 'data/world_bank_projects.json' and the techniques demonstrated above,
1. Find the 10 countries with most projects
2. Find the top 10 major project themes (using column 'mjtheme_namecode')
3. In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.


### Import needed packages

```python
import pandas as pd
import json
from pandas.io.json import json_normalize
```

```python
# load data as Pandas dataframe
dataset = pd.read_json('data/world_bank_projects.json')
```

```python
# Evaluate basic shape of dataset
dataset.shape
```

```python
# Look at example rows of dataset
dataset.head()
```

```python
# Get an idea of completeness of data set
dataset.info()
```

### Answer Question 1: Find the 10 countries with most projects

```python
# Determine unique number of countries and projects
dataset[['countryname', 'project_name']].nunique()
```

```python
# Answer question 1 by grouping project counts by country
Top10 = dataset.groupby('countryname')['project_name'].count().sort_values(ascending=False)
Top10.head(10)
```

### Answer Question 2: Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
#Project themes are nested ata that need to be extracted and then normalized.
dataset2 = json.load((open('data/world_bank_projects.json')))
project_themes = json_normalize(dataset2, 'mjtheme_namecode')
project_themes
```

```python
# Now to determine the top 10 counts for theme, while removing the missing values from the calculation.
project_themes.name[project_themes.name != ""].value_counts().head(10)
```

### Answer Question 3: In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# Create a dataframe of the themes and their codes
uniquethemes = project_themes.drop_duplicates()
uniquethemes = uniquethemes[uniquethemes.name != ""]
uniquethemes = uniquethemes.reset_index().drop('index', axis=1)
uniquethemes
```

```python
# Merge the two datasets on the code to fill in project names.
project_themes = pd.merge(project_themes, uniquethemes, on='code', suffixes=("_x","")).drop("name_x", axis=1)
project_themes.head()
```

```python
# After completion of this merge and fill-in, what are the top projects?
project_themes.name.value_counts().head(10)
```

#### The counts have changed! 

```python
# One last check for blank values in the names of this updated dataset:
project_themes.name[project_themes.name == ""].count()
```
